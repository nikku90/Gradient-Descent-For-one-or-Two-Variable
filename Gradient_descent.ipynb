```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams["figure.figsize"]=[10,8]
```

```python
def define_func(x):
    return 0.1*x**2+np.sin(0.1*x**2)
```

```python
def derivative_func(x):
    return 0.2*x+0.2*x*np.cos(0.1*x**2)
```

```python
def plot_function(x,points,label):
    y=define_func(x)
    
    plt.plot(x,y)
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.title(label)
    
    plt.plot(points, define_func(points), 'bo')
    
    plt.show()
```

```python
x=np.arange(-5,5,0.1)
plot_function(x,np.array([]),"function_plot")
```

```python
def gradient_descent(derivative_func,starting_point,learning_rate):
    
    # Stoping Critiera 
    checking_Criteria=0.001
    
    # Iteration Counter
    iteriatin_count=1
    
    # first Updated point 
    updated_point=starting_point-learning_rate*derivative_func(starting_point)
    
    #Stor the Updated point
    seq=np.array([])
    seq=np.append(seq,updated_point)
    
    # loop till stoping criteria is true
    while abs(updated_point-starting_point)>checking_Criteria:
        
        starting_point=updated_point
        
        updated_point=starting_point-learning_rate*derivative_func(starting_point)
        
        seq=np.append(seq,updated_point)
        
        iteriatin_count+=1
        
    return updated_point,seq,iteriatin_count
```

```python
grad_output=gradient_descent(derivative_func,5,1)
plot_function(x,grad_output[1],"Gradient Descent with alpha =1 and start point = 5")
```

```python
# Two variable Function
```

```python
def func(x,y):
    return x**2 + y**2 + 1
```

```python
def derivative_fun_x(x):
    return 2*x
```

```python
def derivative_fun_y(y):
    return 2*y
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
```

```python
# Plotting in 3D

def plot_func(x,y,dots_x,dots_y,label):
    
    X,Y=np.meshgrid(x, y)
    
    Z = func(X,Y)
    
    fig = plt.figure(1)
    ax = fig.gca(projection='3d')
    surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.RdBu,linewidth=0, antialiased=False)
    ax.scatter(dots_x,dots_y,func(dots_x,dots_y),cmap="black")

    ax.zaxis.set_major_locator(LinearLocator(10))
    ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))
    fig.colorbar(surf,shrink = 0.7, aspect=5)
    ax.set_title(label)
    plt.show()
```

```python
x = np.arange(-5, 5, 0.25)
y = np.arange(-5, 5, 0.25)
plot_func(x,y,np.array([]),np.array([]),"Two Variable plot")
```

```python
def grad_desc(derivative_fun_x,derivative_fun_y,start_x,start_y,learning_rate):
    
    # Stoping Critiera 
    checking_Criteria=0.001
    
    # Iteration Counter
    iteriatin_count=1
    
    
    # first Updated point 
    new_point_x=start_x-learning_rate*derivative_fun_x(start_x)
    new_point_y=start_y-learning_rate*derivative_fun_y(start_y)
    
    sequence_x = np.array([])
    sequence_y = np.array([])
    sequence_x = np.append(sequence_x,new_point_x)
    sequence_y = np.append(sequence_y,new_point_y)
    while abs(np.sqrt((new_point_x-start_x)**2+(new_point_y-start_y)**2))>checking_Criteria:
        
        start_x=new_point_x
        start_y=new_point_y
        new_point_x=start_x-learning_rate*derivative_fun_x(start_x)
        new_point_y=start_y-learning_rate*derivative_fun_y(start_y)
        sequence_x = np.append(sequence_x,new_point_x)
        sequence_y=np.append(sequence_y,new_point_y)
        
        iteriatin_count+=1
    return new_point_x,new_point_y,sequence_x,sequence_y,iteriatin_count
```
